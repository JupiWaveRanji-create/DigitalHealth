```python
# (Optional but recommended) reset from file to undo any earlier overwrites
# df_ambulance = pd.read_csv("ambulance_queue_snapshots_synthetic.csv")

s = (df_ambulance["snapshot_datetime_raw"]
     .astype("string")
     .str.strip()
     .str.replace("+1100", "", regex=False)
)

# Identify AU-style dates like 01/09/2025 00:30
mask_au = s.str.match(r"^\d{2}/\d{2}/\d{4}")

dt = pd.Series(pd.NaT, index=df_ambulance.index, dtype="datetime64[ns]")

# Parse ISO-ish (YYYY-... or YYYY/...) with dayfirst=False
dt.loc[~mask_au] = pd.to_datetime(s.loc[~mask_au], errors="coerce", dayfirst=False)

# Parse AU (DD/MM/YYYY ...) with dayfirst=True
dt.loc[mask_au] = pd.to_datetime(s.loc[mask_au], errors="coerce", dayfirst=True)

df_ambulance["snapshot_datetime_clean"] = dt

# Check NaT rate
df_ambulance["snapshot_datetime_clean"].isna().mean()
```

```python
bad = df_ambulance.loc[df_ambulance["snapshot_datetime_clean"].isna(), "snapshot_datetime_raw"]
bad.value_counts().head(20)
```

```python
s = (
    df_ambulance["snapshot_datetime_raw"]
      .astype(str)
      .str.strip()
      .str.replace("+1100", "", regex=False)
      .str.replace(r"\s+", " ", regex=True)     # collapse multiple spaces
      .str.replace("/", "-", regex=False)       # unify separators
      .str.replace("T", " ", regex=False)       # replace ISO 'T' with space
)

# AU-style now looks like DD-MM-YYYY ...
mask_au = s.str.match(r"^\d{2}-\d{2}-\d{4}")

dt = pd.Series(pd.NaT, index=df_ambulance.index, dtype="datetime64[ns]")

# ISO: YYYY-MM-DD ...
dt.loc[~mask_au] = pd.to_datetime(s.loc[~mask_au], errors="coerce", dayfirst=False)

# AU: DD-MM-YYYY ...
dt.loc[mask_au] = pd.to_datetime(s.loc[mask_au], errors="coerce", dayfirst=True)

df_ambulance["snapshot_datetime_clean"] = dt

df_ambulance["snapshot_datetime_clean"].isna().mean()
```

```python
df_ambulance = df_ambulance.dropna(subset=["snapshot_datetime_clean"])
```

```python
s = (
    df_ambulance["snapshot_datetime_raw"]
      .astype(str).str.strip()
      .str.replace("+1100", "", regex=False)
      .str.replace(r"\s+", " ", regex=True)
      .str.replace("/", "-", regex=False)
      .str.replace("T", " ", regex=False)
)

df_ambulance["snapshot_datetime_clean"] = pd.to_datetime(
    s,
    errors="coerce",
    format="mixed"   # <-- KEY
)

df_ambulance["snapshot_datetime_clean"].isna().mean()
```
